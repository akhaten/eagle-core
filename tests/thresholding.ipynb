```python
import numpy
import matplotlib.pyplot
```

```python
x = numpy.arange(start=-10, stop=10)
```

## Soft thresholding

$\begin{align} 
Soft_{\alpha}(x) 
&= sign(x) \times max(|x| - \alpha, ~0) \\
&= \begin{cases}
~ -x & if  & ~ x \lt -\alpha \\
~ ~ ~ 0  & if & ~ -\alpha \lt x \lt \alpha \\
~ ~ ~ x & if & ~ \alpha \lt x
\end{cases}
\end{align}$

```python
alpha = 5
y = eaglecore.thresholding.soft(x=x, threshold=alpha)
```

```python
fig = matplotlib.pyplot.figure()
matplotlib.pyplot.grid(visible=True)
matplotlib.pyplot.plot(x, y, marker='x')
matplotlib.pyplot.xlabel(xlabel='x')
matplotlib.pyplot.ylabel(ylabel='y')
matplotlib.pyplot.title(label = '$y = Soft_{}(x)$'.format(alpha))
matplotlib.pyplot.show()
matplotlib.pyplot.close()
```
